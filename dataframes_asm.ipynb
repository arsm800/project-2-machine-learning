```python
# Create datetime index conversion function

def convert_datetime(income_statement_df):
    
    # Remove quarters
    dates = [income_statement_df.index[i][:10] for i in range(len(income_statement_df.index))]
    income_statement_df['Date'] = dates
    income_statement_df = income_statement_df.set_index('Date')
    income_statement_df.index = pd.to_datetime(income_statement_df.index, format='%Y-%m-%d')
    
    # Add timestamp
    times = [income_statement_df.index[i] + pd.DateOffset(hours=16) for i in range(len(income_statement_df.index))]
    income_statement_df['times'] = times
    income_statement_df = income_statement_df.set_index('times')
    
    return income_statement_df
```

```python
# Convert income statement dates to datetime

amzn_indexed = convert_datetime(amzn_inc_stmnt)
att_indexed = convert_datetime(att_inc_stmnt)
boa_indexed = convert_datetime(boa_inc_stmnt)
coke_indexed = convert_datetime(coke_inc_stmnt)
csx_indexed = convert_datetime(csx_inc_stmnt)
jpm_indexed = convert_datetime(jpm_inc_stmnt)
odfl_indexed = convert_datetime(odfl_inc_stmnt)
pg_indexed = convert_datetime(pg_inc_stmnt)
tsla_indexed = convert_datetime(tesla_inc_stmnt)
vz_indexed = convert_datetime(vz_inc_stmnt)

#dates = [csx_inc_stmnt.index[i][:10] for i in range(len(csx_inc_stmnt.index))]
#dates

```
